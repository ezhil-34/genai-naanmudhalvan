```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# DCGAN Generator
class Generator(nn.Module):
    def __init__(self, z_dim=100):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.ConvTranspose2d(z_dim, 256, 4, 1, 0, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),

            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),

            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),

            nn.ConvTranspose2d(64, 1, 4, 2, 1, bias=False),
            nn.Tanh()  # Output image range is [-1, 1]
        )

    def forward(self, input):
        return self.main(input)

# DCGAN Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Conv2d(1, 64, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(64, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(128, 256, 4, 2, 1, bias=False),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(256, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()  # Output probability
        )

    def forward(self, input):
        return self.main(input)

# Function to generate random images using the generator
def generate_images(generator, z_dim=100, num_images=1):
    z = torch.randn(num_images, z_dim, 1, 1)  # Latent vector
    with torch.no_grad():
        generated_images = generator(z)
    return generated_images

# Function to display generated images
def show_generated_images(images):
    images = images.squeeze(0).detach().cpu().numpy()
    images = (images + 1) / 2  # Denormalize to [0, 1] range
    plt.imshow(images[0], cmap='gray')
    plt.axis('off')
    plt.show()

# Main function to run everything
def main():
    z_dim = 100  # Latent vector size
    generator = Generator(z_dim=z_dim).eval()

    # Generate random images using the generator
    generated_images = generate_images(generator, z_dim)

    # Show the generated image
    show_generated_images(generated_images)

if __name__ == "__main__":
    main()





```

```python
# Importing required libraries
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from scipy.special import softmax
import gradio as gr

# Load pre-trained model and tokenizer
model_name = "cardiffnlp/twitter-roberta-base-sentiment-latest"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Define sentiment analysis function
def sentiment_analysis(text):
    encoded_input = tokenizer(text, return_tensors='pt')
    output = model(**encoded_input)
    scores = output.logits.detach().numpy()[0]
    scores = softmax(scores)
    labels = ['Negative', 'Neutral', 'Positive']
    results = {label: float(score) for label, score in zip(labels, scores)}
    return results

# Create Gradio interface
demo = gr.Interface(
    fn=sentiment_analysis,
    inputs=gr.Textbox(placeholder="Write your text here..."),
    outputs="json",
    examples=[
        ["I'm thrilled about the job offer!"],
        ["The weather today is absolutely beautiful."],
        ["I had a fantastic time at the concert last night."],
        ["I'm so frustrated with this software glitch."],
        ["The customer service was terrible at the store."],
        ["I'm really disappointed with the quality of this product."]
    ],
    title="Sentiment Analysis App",
    description="This app classifies text into positive, neutral, or negative sentiment."
)

# Launch the app
demo.launch()
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from scipy.special import softmax
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
def analyze_sentiment(text):
    encoded_input = tokenizer(text, return_tensors='pt')
    output = model(**encoded_input)
    scores = output.logits.detach().numpy()[0]
    scores = softmax(scores)
    labels = ['Negative', 'Positive']
    results = {label: float(score) for label, score in zip(labels, scores)}
    return results
def plot_sentiment_distribution(results):
    labels = list(results.keys())
    scores = list(results.values())

    plt.figure(figsize=(6, 4))
    plt.bar(labels, scores, color=['red', 'green'])
    plt.title('Sentiment Distribution')
    plt.ylabel('Score')
    plt.ylim(0, 1)
    plt.savefig('sentiment_distribution.png')

    return 'sentiment_distribution.png'

demo = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.Textbox(placeholder="Enter social media text here..."),
    outputs=["json", gr.Image(type="filepath")],
    title="Social Media Sentiment Analysis",
    description="Analyze sentiments of social media posts."
)

# Launch the app
demo.launch()
```

```python
import gradio as gr
from transformers import pipeline
import numpy as np

# Load model on CUDA device
transcriber = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-base.en",
    device="cuda:0"
)

def transcribe(audio):
    sr, y = audio
    if y.ndim > 1:  # Convert stereo to mono
        y = y.mean(axis=1)
    y = y.astype(np.float32)
    y /= np.max(np.abs(y))  # Normalize
    return transcriber({"sampling_rate": sr, "raw": y})["text"]

# Corrected Audio component with 'sources' parameter
demo = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(sources="microphone", type="numpy"),  # Changed source->sources
    outputs="text"
)

demo.launch()
```

```python
import gradio as gr
from transformers import pipeline

# Load the question-answering model
qa_model = pipeline("question-answering")

# Define the Q&A function
def answer_question(question, context):
    result = qa_model(question=question, context=context)
    return result['answer']

# Create a Gradio interface
demo = gr.Interface(
    fn=answer_question,
    inputs=[
        gr.Textbox(label="Question", placeholder="Enter your question here..."),
        gr.Textbox(label="Context", placeholder="Provide context for the question...")
    ],
    outputs="text"
)

# Launch the app
demo.launch()
```

```python
 import gradio as gr
from transformers import pipeline

# Load the summarization model
summarizer = pipeline("summarization")

# Define the summarization function
def summarize_text(text):
    summary = summarizer(text, max_length=130, min_length=30, do_sample=False)
    return summary[0]['summary_text']

# Create a Gradio interface
demo = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(label="Input Text", placeholder="Enter text to summarize...", lines=10),
    outputs="text"
)

# Launch the app
demo.launch()
```

```python
import gradio as gr
from transformers import pipeline

# Load the text generation model
generator = pipeline("text-generation", model="gpt2")

# Define the paragraph generation function
def generate_paragraph(prompt):
    result = generator(prompt, max_length=200, num_return_sequences=1, do_sample=True)
    return result[0]['generated_text']

# Create a Gradio interface
demo = gr.Interface(
    fn=generate_paragraph,
    inputs=gr.Textbox(label="Input Prompt", placeholder="Enter a topic or starting sentence...", lines=2),
    outputs=gr.Textbox(label="Generated Paragraph", lines=10),
    title="Text Paragraphing with AI",
    description="Generate well-structured paragraphs based on your input prompt using GPT-2."
)

# Launch the app
demo.launch()
```

```python
import gradio as gr
from gtts import gTTS
import os

# Define the text-to-speech function
def text_to_speech(text):
    tts = gTTS(text, lang="en")  # Convert text to speech
    output_file = "output.mp3"
    tts.save(output_file)  # Save audio to file
    return output_file  # Return the audio file path

# Create a Gradio interface
demo = gr.Interface(
    fn=text_to_speech,
    inputs=gr.Textbox(label="Input Text", placeholder="Enter text to convert to speech..."),
    outputs=gr.Audio(label="Generated Speech"),
    title="Text-to-Speech App",
    description="Convert text into speech using the gTTS library."
)

# Launch the app
demo.launch()
```
